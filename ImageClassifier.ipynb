## CNN Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()

model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
opz = tf.optimizers.Adam(learning_rate=0.001) # this affected greatly 
```

```python
model.compile(optimizer= opz,
             loss= tf.losses.SparseCategoricalCrossentropy(),
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
## Training
logdir = 'logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logdir)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
hist = model.fit(train,epochs=20, 
                 validation_data= val,
                 callbacks= [tensorboard_callback])
```

```python
# Plot training loss
plt.subplot(1, 2, 1)
plt.plot(hist.history['loss'], label='Training Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

    # Plot training accuracy
plt.subplot(1, 2, 2)
plt.plot(hist.history['accuracy'], label='Training Accuracy')
plt.plot(hist.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

    # Adjust layout for better visualization
plt.tight_layout()

    # Show the plots
plt.show()
```

```python
last_epoch_results = {
        'last_epoch_train_loss': hist.history['loss'][-1],
        'last_epoch_val_loss': hist.history['val_loss'][-1],
        'last_epoch_train_accuracy': hist.history['accuracy'][-1],
        'last_epoch_val_accuracy': hist.history['val_accuracy'][-1],
    }

    # Print last epoch metrics
print("\nLast Epoch Metrics:")
print(f"Train Loss: {last_epoch_results['last_epoch_train_loss']:.4f}")
print(f"Validation Loss: {last_epoch_results['last_epoch_val_loss']:.4f}")
print(f"Train Accuracy: {last_epoch_results['last_epoch_train_accuracy']:.4f}")
print(f"Validation Accuracy: {last_epoch_results['last_epoch_val_accuracy']:.4f}")
```

## Evaluation

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
from tensorflow.keras.utils import to_categorical

precision = Precision()
recall = Recall()
accuracy = CategoricalAccuracy()

for batch in test.as_numpy_iterator():
    X, y = batch  # the y here is the true value label
    y_true_one_hot = to_categorical(y, num_classes=3)
    y_pred = model.predict(X)
    y_pred = y_pred.reshape(y.shape[0], 3)

    precision.update_state(y_true_one_hot, y_pred)
    recall.update_state(y_true_one_hot, y_pred)
    accuracy.update_state(y_true_one_hot, y_pred)
```

```python
print(f'Precision:{precision.result().numpy()}, Recall:{recall.result().numpy()},Accuracy:{accuracy.result().numpy()}')
```
